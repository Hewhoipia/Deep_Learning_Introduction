#1 Branching: if statement



##1.1 Samples

###1.1.1 if statement

```python
import math
x = int(input("Enter an integer:"))
print(f'x = {x}')
if x >= 0:
  print('x >= 0; SQRT(x) = {:<10.2f}'.format(math.sqrt(x)))
```

###1.1.2 if-else statement


```python
valid_low, valid_high = 0, 10
print(f'valid range: [{valid_low}, {valid_high}]')
x = int(input("Enter an integer:"))
if (x >= valid_low) & (x <= valid_high):
  print(f'x = {x}, a valid value')
else:
  print(f'x = {x}, it is not inside of the valid range')
```

###1.1.3 else-if statement

```python
valid_low, valid_high = 0, 10
print(f'valid range: [{valid_low}, {valid_high}]')
x = int(input("Enter an integer:"))
if x < valid_low:
  print(f'x = {x}, it is smaller than the lower-bound; so, it is invalid')
elif x > valid_high:
  print(f'x = {x}, it is larger than the upper-bound; so, it is invalid')
else:
  print(f'x = {x}, a valid value')
```

###2.1.1 for with function range
Notes:
* ***range(start=0, stop, step=1)***: generates a sequence from ***start*** (included)to ***stop***(excluded), increments by ***step***,



```python
#start is 0 by default, stop is 10 (not included 10), increment=1 by default
for v in range(10):
  print(v)
```

```python
#start is 2, stop is 10 (not included 10), increment=1 by default
for v in range(2, 10):
  print(v)
```

```python
#start is 2, stop is 10 (not included 10), increment=2

for v in range(2, 10, 2):
  print(v)
```

####Q.2.2 Must use comprehension

```python
"""
Fill code to this cell to compute the sum of all the numeric values in list L,
and show the result as in the output cell followed
"""
L = [36, 2, 27, 'ABC', 30, 36, 0, 13, 29, 'None', 31, 17]
#YOUR CODE IS HERE
L1 = [v for v in L if isinstance(v, (int, float))]
total = sum(L1)

print("List: ", L)
print("Total:", total)
```

###Q.3

####Q.3.1 Must use for-loop

```python
"""
Fill code to this cell to compute the average of all the numerical values in list L,
and show the result as in the output cell followed
"""
L = [36, 2, 27, 'ABC', 30, 36, 0, 13, 29, 'None', 31, 17]
#YOUR CODE IS HERE
total = 0
count = 0
for e in L:
  if isinstance(e, (int, float)):
    total += e
    count += 1
average = total/count
print("List: ", L)
print("Total:", total)
print("Average: {:<7.2f}".format(average))
```

####Q.3.2 Must use comprehension

```python
"""
Fill code to this cell to compute the average of all the numerical values in list L,
and show the result as in the output cell followed
"""
L = [36, 2, 27, 'ABC', 30, 36, 0, 13, 29, 'None', 31, 17]
#YOUR CODE IS HERE
L1 = [v for v in L if isinstance(v, (int, float))]
total = sum(L1)
average = total/len(L1)
print("List: ", L)
print("Total:", total)
print("Average: {:<7.2f}".format(average))
```
