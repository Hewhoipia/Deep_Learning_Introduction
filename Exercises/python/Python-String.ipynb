##1.1 Split a string to items

```python
# Split string with token = " "
course = "Programming Foundation for Data Science"
L = course.split(" ")
L
```

```python
# Split string with token = " "
course = "Programming Foundation ; for ; Data Science"
L = course.split(";")
print(L)
L = [s.strip() for s in L]
print(L)
```

##1.2 string to list/tuple/set

```python
s = "ABCDE"
L = list(s)
print("s = ", s)
print("L = ", L)
```

```python
s = "ABCDE"
L = tuple(s)
print("s = ", s)
print("L = ", L)
```

```python
s = "ABCDE"
L = set(s)
print("s = ", s)
print("L = ", L)
```

##1.3 Join items to string

```python
L = ['Programming', 'Foundation', 'for', 'Data Science']
s = ' '.join(L)
print("L = ", L)
print("s = ", s)
```

```python
L = [10, 2, 15, 7, 8, 25]
s = "[" + ','.join([str(item) for item in L]) + "]"
print("L = ", L)
print("s = ", s)
```
