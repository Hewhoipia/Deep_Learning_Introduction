```python
import os
import math
import functools as F
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow.keras.utils as kutils
```

```python
def show_image(image, title):
  plt.figure(figsize=(5,5))
  plt.imshow(image)
  plt.axis("off")
  plt.title(title)

image_file = './images/Abyssinian_104.jpg'
label_file = './annotations/trimaps/Abyssinian_104.png'
shape = (200,200)

image = np.asarray(kutils.load_img(image_file, target_size=shape, color_mode='rgb'))
label = np.asarray(kutils.load_img(label_file, target_size=shape, color_mode='grayscale'))
```

```python
print("Type: ", image.dtype, label.dtype)
print("Shape: ", image.shape, label.shape)
print("Dimensions: ", image.ndim, label.ndim)
print("Max, min: ", max(image[:,0,0]), min(image[:,0,0]), max(label[0]), min(label[0]))
```

```python
newImage=np.copy(image)
mask=(label!=1)
newImage[mask]=[100,100,100]
show_image(newImage,"1")
```
