## Q.2 - Defining/Visulizing 3D-Mesh using **np.meshgrid**
* Requirements:
  * Create a 3D-mesh as shown in the figure below
* Guidelines:
  * See the example: [https://matplotlib.org/stable/gallery/mplot3d/surface3d.html](https://matplotlib.org/stable/gallery/mplot3d/surface3d.html)

```python
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator
import numpy as np

#YOUR CODE IS HERE
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
x=np.arange(-5,5,0.1)
y=np.arange(-5,5,0.1)
x, y = np.meshgrid(x, y)
z=x**2 + y**2
surf = ax.plot_surface(x, y, z, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)
# Customize the z axis.
ax.set_zlim(0.00, 50.00)
ax.zaxis.set_major_locator(LinearLocator(10))
# A StrMethodFormatter is used automatically
ax.zaxis.set_major_formatter('{x:.02f}')

# Add a color bar which maps values to colors.
fig.colorbar(surf, shrink=0.5, aspect=5)
plt.title(r"$z=x^2 + y^2$")
plt.show()
```

## Q.2- Sampling 2D-normal

### Q.2.1 Examples

```python
# Declare variables
mu = (5, 3)
sigma = (3,4)
N = 10000

# Generate data from 2D-normal(mu sigma)
X = np.random.normal(mu, sigma, (N,2)) #N rows, 2 cols

# Show the 2D-histogram
fig = plt.figure(figsize=(6,4))
ax = fig.gca() # you can get axes from plt.subplot(,,)
_, _, _, im = ax.hist2d(X[:, 0], X[:, 1], bins=10*10, density=True)
title = "Histogram from {} points".format(N)
ax.set_xlabel('$x_1$')
ax.set_ylabel('$x_2$')
ax.set_title(title)
fig.colorbar(im, ax=ax)
plt.show()
```

## Q.2 - Estimating the distribution's parameters from the generated dataset
* Requirements:
  * Fill code to the cell followed to estimated values and their probabilities from the generated data.
    * Guidelines:
      * Use ``values, counts = np.unique(..., return_counts=True)``
        * Link: [https://numpy.org/doc/stable/reference/generated/numpy.unique.html](https://numpy.org/doc/stable/reference/generated/numpy.unique.html)
      * Use ``np.argsort`` to sort ``values`` and ``counts``
      * Compute the probabilty from ``counts`` by dividing it for ``counts.sum()``
  * Print the report as shown in the output


```python
animals =     ['Cat', 'Chicken', 'Dog', 'Horse']
probability =  [0.1,   0.4,        0.2,   0.3] # must sum to 1
N = 10000 # number of generated animals

# YOUR CODE IS HERE
array=np.random.choice(animals, size=(N,), p=probabilty)
values, counts=np.unique(array, return_counts=True)
ev=values
ep=np.array([c/counts.sum() for c in counts])
proba=np.array(probability)
error=np.abs(proba - ep)
print("====================================================================================================")
print(f"Number of samples (N): {N}")
print("Input values (animals): ",animals)
print("Input probability (probability) ",probability)
print("--------------------------------------------------")
print("Estimated values: ",ev)
print("Estimated probability ",ep)
print("--------------------------------------------------")
print("ABS(Errors): ",error)
```
