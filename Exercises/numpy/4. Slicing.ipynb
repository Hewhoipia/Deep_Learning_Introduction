# 1- Slicing - (I)

## 1.1- Examples
* Syntax:
  * use ```':'``` to select all indices at one dimension
  * use ```'...'``` to select all the indices in all the remaining dimensions

````python
# declare variables
N, M = 6, 4

# generate N integers
np.random.seed(0)
X = np.random.randint(0, 10, (N,M))

# select items using their indices
np.set_printoptions(formatter={'float': '{:<0.3f}'.format,
                                'int': '{:<3d}'.format,
                                'bool': '{:<3}'.format })
print("X: \t"); print(X)
print('-'*150)

print('To select colum 0, X[:, 0]: ', X[:, 0])
print('To select colum 0, X[..., 0]: ', X[..., 0])
print('To select colum 2, X[:, 2]: ', X[:, 2])
print('To select colum 2, X[..., 2]: ', X[..., 2])
print('-'*50)
print('To select colum 0 and 2, X[:, [0, 2]]: '); print(X[:, [0,2]])
print('To select colum 0 and 2, X[:, (0, 2)]: '); print(X[:, (0,2)])
print('To select colum 0 and 2, X[..., [0, 2]]:'); print(X[..., [0,2]])
print('To select colum 0 and 2, X[..., (0, 2)]:'); print(X[..., (0,2)])
````

## Question-1.
* Requirements:
  1. Select the items in columns with even indices
  2. Select the items in rows with odd indices
  3. Show the results as in the output cell

````python
# declare variables
N, M = 10, 8

# generate N integers
np.random.seed(0)
X = np.random.randint(0, 10, (N,M))

# select items using their indices
np.set_printoptions(formatter={'float': '{:<0.3f}'.format,
                                'int': '{:<3d}'.format,
                                'bool': '{:<3}'.format })
print("X: \t"); print(X)
print('-'*150)

#YOUR CODE IS HERE
print('Items in even collumns:'); print(X[:, [idx for idx in range(M) if (idx%2==0)]])
print('Items in odd rows:'); print(X[[idx for idx in range(M) if (idx%2!=0)],:])
````
